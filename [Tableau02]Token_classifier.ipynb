# Utility Functions

```python
def master_token_allocator(df, token_column_name, similar_token, pointers):
    
    for index in range(0, len(df)):
        
        index_data = df.loc[index]
        token = index_data[token_column_name]
        ticker = 0
        
        for master_word in similar_token:
            if token in similar_token[master_word]:
                df.loc[index, 'master_word'] = master_word
                ticker = 1
            else:
                ticker = 0
                pass
        
        if ticker == 0:
            
            for pointer in pointers:
                
                if token in pointers[pointer]:
                    df.loc[index, 'master_word'] = token
                    ticker = 1
                
                else:
                    ticker = 0
                    continue
    
    return df
                
```

```python
def category_allocator(df, master_token_column_name, pointers):
    
    for index in range(0, len(df)):
        
        index_data = df.loc[index]
        mtoken = index_data[master_token_column_name]
        ticker = 0
        
        for pointer in pointers:
            
            if mtoken in pointers[pointer]:
                df.loc[index, 'token_category'] = pointer
            
            elif mtoken not in pointers[pointer]:
                continue
    #df.fillna('uncategorized')
    return df
```
