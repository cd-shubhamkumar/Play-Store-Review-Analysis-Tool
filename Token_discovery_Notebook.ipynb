# **Defining Utility Functions**

```python
def word_dataframe(input_df, review_column, upvote_column, rating):
    #user_reviews_column = input_df[review_column]
    #upvote = input_df[upvote_column]
    
    df = pd.DataFrame(columns = ['word_token', 'upvote_sum', 'rating', 'word_rep_count', 'user_count'])
    
    for review in range(0,len(input_df)):
        list_of_tokens = word_tokenize(input_df.iloc[review][review_column])
        upvotes = input_df.iloc[review][upvote_column]
        
        for token_number in range(0, len(list_of_tokens)):
            
            token = list_of_tokens[token_number]
            
            if len(df[df['word_token'] == token]) == 0:
                df = df.append({'word_token':token, 
                                'upvote_sum':upvotes, 
                                'rating':rating, 
                                'word_rep_count':1, 
                                'user_count':1}, ignore_index = True)
            
            elif len(df[df['word_token'] == token]) == 1:
                index = df[df['word_token'] == token].index
                df.loc[index,'upvote_sum'] = df.loc[index,'upvote_sum'] + upvotes
                df.loc[index,'word_rep_count'] = df.loc[index,'word_rep_count'] + 1
                if token not in list_of_tokens[0:token_number]:
                    df.loc[index,'user_count'] = df.loc[index,'user_count'] + 1
                elif token in list_of_tokens[0:token_number]:
                    pass
            
            else:
                continue
        
    return df
                
```

```python
def remove_single_char_token(df, token_column):
    
    df_index_list = df.index.to_list()
    output_index_list = []
    
    for df_index_num in range(0, len(df_index_list)):
        df_index = df_index_list[df_index_num]
        
        if len(df.loc[df_index][token_column]) > 2:
            output_index_list.append(df_index_num)
        
        else:
            continue
    
    output_df = df.iloc[output_index_list]
    
    return output_df
```

```python
def remove_longer_token(df, token_column):
    
    df_index_list = df.index.to_list()
    output_index_list = []
    
    for df_index_num in range(0, len(df_index_list)):
        df_index = df_index_list[df_index_num]
        
        if len(df.loc[df_index][token_column]) < 17:
            output_index_list.append(df_index_num)
        
        else:
            continue
    
    output_df = df.iloc[output_index_list]
    
    return output_df
```

```python
def print_distinct_words(input_df, column_name, column_name2):
    
    word_column = []
    word_column.append(np.array(input_df[column_name]))
    word_column.append(np.array(input_df[column_name2]))
    print('Count of all Words ---> \t', len(word_column[0]), '\n')
    
    for word in range(0, len(word_column[0])):
        if len(word_column[0][word]) > 7:
            print(word_column[0][word], '\t\t', word_column[1][word])
        elif len(word_column[0][word]) < 7:
            print(word_column[0][word], '\t\t\t', word_column[1][word])
```

```python
def save_distinct_words(input_df, column_name):
    
    word_column = input_df[column_name]
    output = []
    
    for word in word_column:
        output.append(word)
    
    return output
```

```python
def token_dictionary(list_of_token_list):
    
    number_of_lists = len(list_of_token_list)
    output = []
    
    for review_list in list_of_token_list:
        for token in review_list:
            
            if token in output:
                continue
            
            elif token not in output:
                output.append(token)
            
            else:
                pass
    
    return output
```

```python
def token_dictionary_with_ratings(dictionary_of_rating_df, token_column_name):
    
    output_token_dictionary = {}
    
    for rating in dictionary_of_rating_df:
        token_list = dictionary_of_rating_df[rating][token_column_name]
        
        for token in token_list:
            
            if token not in output_token_dictionary:
                output_token_dictionary[token] = [rating]
            
            elif token in output_token_dictionary:
                output_token_dictionary[token].append(rating)
            
            else:
                continue
    
    return output_token_dictionary
```
