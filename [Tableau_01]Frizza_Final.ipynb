```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import nltk
from nltk.tokenize import word_tokenize 
nltk.download('punkt')
```

# **Importing Data**

```python
token = pd.read_csv('token_list.csv')
token_list = np.array(token['token']).tolist()
token_list
```

```python
data = pd.read_csv('frizza_reviews.csv')
reviews = pd.DataFrame(data)
reviews['at'] = pd.to_datetime(reviews['at'])
reviews['repliedAt'] = pd.to_datetime(reviews['repliedAt'])
reviews['date'] = reviews['at'].dt.date

##Converting all text into lower case
for review_text_index in range(0, len(reviews)):
    rev_text = reviews.loc[review_text_index]['content']
    reviews['content'] = reviews['content'].replace([rev_text], rev_text.lower())
   
    
reviews.tail()
```

# **Dividing the DataFrame w.r.t. Ratings**

```python
rdf5 = reviews[reviews['score'] == 5]
rdf5 = rdf5.reset_index()

rdf4 = reviews[reviews['score'] == 4]
rdf4 = rdf4.reset_index()

rdf3 = reviews[reviews['score'] == 3]
rdf3 = rdf3.reset_index()

rdf2 = reviews[reviews['score'] == 2]
rdf2 = rdf2.reset_index()

rdf1 = reviews[reviews['score'] == 1]
rdf1 = rdf1.reset_index()
```

# **Utility Functions**

```python
def df_to_dictionary(df, date_column_name, version_column_name, rating, review_column_name, token_dictionary):
    
    output = {}
    #rating = ratings_column_name
    
    for index in range(0, len(df)):
        
        row_data = df.loc[index]
        
        token_list = word_tokenize(row_data[review_column_name])
        date = row_data[date_column_name]
        version = row_data[version_column_name]
        
        for token in token_list:
            
            if token in token_dictionary:
            
                if token not in output:

    #                 version_dict = {version : 1}
    #                 date_dict = {date : version_dict}

                    output[token] = {date : {version : 1}}


                elif token in output:

                    if date in output[token]:
                        #
                        if version in output[token][date]:
                            #
                            output[token][date][version] += 1


                        elif version not in output[token][date]:
                            #
                            #version_dict = {version : 1}
                            output[token][date][version] = 1

                    elif date not in output[token]:
                        #
    #                     version_dict = {version : 1}
    #                     date_dict = {date : version_dict}
                        output[token][date] = {version : 1}
            else:
                pass
    
    return output
```

```python
def dictionary_to_df(dictionary, rating):
    
    df = pd.DataFrame(columns = ['token', 'rating', 'date', 'version', 'used_count'])
    
    for token in dictionary:
        
        if (len(token) > 2) and (len(token) < 17):
            for date in dictionary[token]:
            
                for version in dictionary[token][date]:
                    used_count = dictionary[token][date][version]
                    input_dict = {'token': token, 'rating': rating, 'date': date, 'version': version, 'used_count': used_count}

                    df = df.append(input_dict, ignore_index = True)
        else:
            pass
        
                
    return df
```

# **For Different Ratings**

```python
date_column_name = 'date'
version_column_name = 'reviewCreatedVersion'
rating = 5
review_column_name = 'content'
token_dictionary = token_list

dict5 = df_to_dictionary(rdf5, date_column_name, version_column_name, rating, review_column_name, token_dictionary)
dict5
```

```python
df5 = dictionary_to_df(dict5, 5)
df5
```

```python
##Export This DataFrame
df5.to_csv(r'for_tableau_mid/rating5/df5.csv', index = False)
```
